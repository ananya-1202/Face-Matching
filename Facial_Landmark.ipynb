```
distance1 = []
distance2 = []
```

```
for (m,n) in points:
  a1 = l1[m-1]
  a2 = l2[m-1]
  b1 = l1[n-1]
  b2 = l2[n-1]
  distance1.append(btwdis(a1,b1))
  distance2.append(btwdis(a2,b2))
```

```
ratio1 = []
ratio2 = []
```

```
for i in range(len(distance1)):
  for j in range(len(distance1)-1-i):
    ratio1.append(distance1[i]/distance1[j+i+1])
    ratio2.append(distance2[i]/distance2[j+i+1])
```

```
sum = 0 
for i in range(len(ratio1)):
  print((((ratio1[i] - ratio2[i])**2)**0.5)/(ratio1[i] + ratio2[i])**2)
  sum = sum + (((ratio1[i] - ratio2[i])**2)**0.5)/(ratio1[i] + ratio2[i])**2
```
